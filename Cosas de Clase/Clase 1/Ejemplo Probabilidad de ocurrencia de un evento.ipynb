## Ejemplo de cálculo de probabilidad por frecuencia relativa del resultado de un evento
Utilizando el concepto de frecuencia de ocurrencia relativa para resolver problemas prácticos de probabilidad, realizar un código en Python que resuelva la siguiente consigna por método Monte Carlo: 

¿ Cuál es la probabilidad de que aparezca el número seis en el lanzamiento de un dado?

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
numIter = np.arange(100,20000,200)
Pa =np.zeros(len(numIter))
k = 0
for veces in numIter:
    Na = 0
    dado = np.zeros(veces)
    for N in range(veces):
        dado[N] = random.randint(1,6)
        if dado[N] == 6:
            Na+=1 
    Pa[k]= Na/N
    k+=1
```

```python
plt.plot(numIter,Pa)
plt.xlabel('Número de iteraciones')
plt.ylabel('Na/N')
plt.title('Cálculo de la probabilidad por frecuencia relativa')
plt.grid()
plt.show()
```

```python
print("La ultima probabilidad calculada es de: ",Pa[-1])
```
