$f_Z(z)= \frac{1}{2}e^{-|z- \mu|}$    , donde  $\mu$  es la media de la variable  Z .

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2 # Se define a como un entero positivo
z = np.arange(-5,5,.5) 
fz1 = 0.5*np.exp(-np.abs(z-a)) #Funciones laplacianas
fz2 = 0.5*np.exp(-np.abs(z+a))

plt.grid()
plt.ylabel("Probabilidad")
plt.xlabel("Observable 'y'")
plt.title("Probabilidad f(Y/H)")
plt.plot(z,fz1,color = "black")
plt.plot(z,fz2,color = "red")
plt.show()
```
